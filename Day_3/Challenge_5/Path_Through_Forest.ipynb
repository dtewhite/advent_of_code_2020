```python
import pandas as pd
import math
```

```python
forest_df = pd.read_csv("forest.csv")
```

```python
forest_df.head()
```

```python
course_len = len(forest_df)
course_len
```

```python
df_width = len(forest_df.iloc[0,:])
df_width
```

```python
total_width_needed = course_len*3
total_width_needed
```

```python
num_of_df_needed = total_width_needed/df_width
num_of_df_needed
```

```python
for i in range(int(math.ceil(math.sqrt(num_of_df_needed)))):
    if len(forest_df.columns) >total_width_needed:
        break
    else:
        forest_df = pd.concat([forest_df,forest_df], axis=1)
```

```python
lateral = 3
vertical = 1
tree_count = 0

for i in range(len(forest_df)-1):
    if forest_df.iloc[vertical,lateral] == "#":
        tree_count += 1
    lateral = lateral + 3
    vertical = vertical + 1
```
